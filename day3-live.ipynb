```python
with open('all_species.txt', 'r', encoding = 'utf-8') as infile:
    names = infile.read().splitlines()
```

```python
len(names) # should be 29814
```

```python
names
```

```python
import re
```

```python
match = re.compile("(.+)\(([0-9]+)\)")

results = []

for n in names:
    one_match = re.findall(match, n)
    results.append(one_match[0])
```

```python
species_pics = {name: int(count) for name, count in results }
```

```python
species_pics
```
